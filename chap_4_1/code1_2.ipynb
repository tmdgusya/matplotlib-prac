```python
import os
import matplotlib.pyplot as plt

img_dir = 'images'
os.makedirs(img_dir, exist_ok=True)

img_file_name = 'tmp.png'

fig = plt.figure(facecolor='skyblue', edgecolor='b', linewidth=3)
ax = fig.add_subplot()

fig_patch = fig.patch
fig_patch.set_linestyle('--')
fig_patch.set_hatch('*')

plt.show()

fig.savefig(os.path.join(img_dir, img_file_name))
```
