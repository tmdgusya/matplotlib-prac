```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=2)

print(type(axes)) # <class 'numpy.ndarray'>
print(axes.shape) # (2,)

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2)
print(type(axes))
print(axes.shape)
print(axes)

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2)

for axes_row in axes:
    for ax in axes_row:
        print(ax)

for ax in axes.flat:
    print(ax)

plt.show()
```
