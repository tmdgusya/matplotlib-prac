```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 100)
y1 = x**2
y2 = (x-1)**2
y3 = (x+1)**2

# add legend
plt.plot(x, y1, label='$y=x^2$')
plt.plot(x, y2, label='$y=(x-1)^2$')
plt.plot(x, y3, label='$y=(x+1)^2$')
plt.legend(fontsize=12)

plt.show()
```

```python
std = 0.5
rng = np.random.default_rng(0)

x1_1 = rng.normal(1, std, 100)
y1_1 = rng.normal(1, std, 100)

x1_2 = rng.normal(-1, std, 100)
y1_2 = rng.normal(-1, std, 100)

x2_1 = rng.normal(-1, std, 100)
y2_1 = rng.normal(1, std, 100)

x2_2 = rng.normal(1, std, 100)
y2_2 = rng.normal(-1, std, 100)

plt.scatter(x1_1, y1_1, label='class 1')
plt.scatter(x1_2, y1_2, label='class 2')
plt.scatter(x2_1, y2_1, label='class 3')
plt.scatter(x2_2, y2_2, label='class 4')
plt.legend()
plt.show()
```
