```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.size'] = 15 

fig, ax = plt.subplots(dpi=200, figsize=(10, 5))

left_loc = [0.01, 0.99]
y_loc = np.linspace(0.05, 0.85, 4)[::-1]
linestyles = ['-', ':', '--', '-.']

for idx in range(len(linestyles)):
    ax.plot(left_loc, [y_loc[idx], y_loc[idx]], linestyle=linestyles[idx], linewidth=1)
    ax.text(left_loc[0], y_loc[idx] + 0.01, f"linestyle='{linestyles[idx]}'", fontsize=10)

ax.axis('off')
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

plt.tight_layout()
plt.show()
```
