```python
import numpy as np
import matplotlib.pyplot as plt

n_data = 300
rng = np.random.default_rng(0)
x = rng.normal(0, 1, size=n_data)
y = rng.normal(0, 1, size=n_data)

plt.scatter(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n_data = 300
rng = np.random.default_rng(0)
X = rng.normal(0, 1, size=(n_data, 2))

plt.scatter(X[:, 0], X[:, 1])
plt.show()
```
