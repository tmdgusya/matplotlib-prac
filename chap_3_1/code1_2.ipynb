```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(dpi=200)

ax.bar([0, 1, 2], [2, 5, 3], facecolor='brown', hatch='\\')

fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(dpi=200)

ax.bar([0, 1, 2], [2, 5, 3], facecolor='green', hatch='**', edgecolor='w')

fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['lines.markersize'] = 15

markers = ['v', '^', '<', '>', 's', 'P', '*', '+', 'x', 'X', 'D', 'd']

x = np.linspace(0, 1, 5)

fig, ax = plt.subplots(dpi=200)

for marker_idx, marker in enumerate(markers):
    y = np.full_like(x, fill_value=-marker_idx)
    ax.scatter(x, y, edgecolors='w', marker=marker)

plt.show()
```
