```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots() # 명시적으로 figure, axes object 를 만들고, 이 object 의 method 를 이용하는 interface

plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure() # create an figure object
ax = fig.add_subplot() # create an axes object in the figure

print(type(fig))
print(type(ax))

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

print(f"{fig.get_dpi()=}")

fig.set_dpi(1000)
print(f"{fig.get_size_inches()=}")

fig.set_size_inches(10, 10)
print(f"{fig.get_facecolor()=}")

fig.set_facecolor("lightgray")
print(f"{fig.get_figwidth()=}")
print(f"{fig.get_figheight()=}")

plt.show()
```
