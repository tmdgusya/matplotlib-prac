```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1) # 2 by 1 중에서 첫번째
plt.plot(x, y_sin, color='red', label='sin')
plt.legend()

plt.subplot(2, 1, 2) # 2 by 1 중에서 두번째 
plt.plot(x, y_cos, color='blue', label='cos')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(1, 2, 1)
plt.plot(x, y_sin, color='red', label='sin')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, y_cos, color='blue', label='cos')
plt.legend()

plt.show()
```

```python
x = np.linspace(-2, 2, 3)
y = x**2

plt.subplot(2, 2, 1)
plt.plot(x, y, color='red', label='y=x^2')
plt.legend()

x = np.linspace(-2, 2, 5)
y = x**2

plt.subplot(2, 2, 2)
plt.plot(x, y, color='red', label='y=x^2')
plt.legend()

x = np.linspace(-2, 2, 11)
y = x**2

plt.subplot(2, 2, 3) # 알아서 줄바꿈 해줌
plt.plot(x, y, color='red', label='y=x^2')
plt.legend()

x = np.linspace(-2, 2, 21)
y = x**2

plt.subplot(2, 2, 4)
plt.plot(x, y, color='red', label='y=x^2')
plt.legend()

plt.show()
```

```python
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y_sin, color='red', label='sin')
plt.title('Sine function', fontsize=15)
plt.xlabel('x', fontsize=13)
plt.ylabel('y', fontsize=13)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x, y_cos, color='blue', label='cos')
plt.title('Cosine function', fontsize=15)
plt.legend()

plt.tight_layout()
plt.show()
```
