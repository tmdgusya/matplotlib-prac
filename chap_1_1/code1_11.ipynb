```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1) # 2 by 1 중에서 첫번째
plt.plot(x, y_sin, color='red', label='sin')
plt.legend()

plt.subplot(2, 1, 2) # 2 by 1 중에서 두번째 
plt.plot(x, y_cos, color='blue', label='cos')
plt.legend()
plt.show()
```
