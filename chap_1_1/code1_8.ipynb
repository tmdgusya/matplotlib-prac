```python
import matplotlib.pyplot as plt

plt.plot([1, 4, 2])

plt.title('Test title', fontsize=20)
plt.xlabel('x axis', fontsize=15)
plt.ylabel('y axis', fontsize=15)
plt.tick_params(labelsize=15)
plt.savefig('tmp.png')
```

```python
plt.plot([1, 4, 2], linewidth=10)
plt.show()
```

```python
plt.plot([1, 4, 2], linewidth=10, linestyle="--")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter([1, 4, 2], [1, 4, 2], s=100, marker="x", color="red")
plt.scatter([0, 3, 6], [1, 4, 2], s=100, marker="o", color="blue")
plt.scatter([1, 4, 2], [0, 3, 6], s=100, marker="^", color="green")
plt.show()
```
