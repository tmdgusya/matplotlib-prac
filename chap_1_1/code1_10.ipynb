```python
import numpy as np
import matplotlib.pyplot as plt

years = list(range(2017, 2024))
real_gdp = [1963, 2026, 2073, 2058, 2153, 2212, 2243]
nominal_gdp = [1943, 2007, 2040, 2058, 2212, 2323, 2401]

plt.bar(years, real_gdp, label='real gdp')
plt.plot(years, nominal_gdp, label='nominal gdp', color='red')
plt.legend()
plt.ylim(bottom=1800, top=2500)
plt.show()
```
