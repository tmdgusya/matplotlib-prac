```python
import matplotlib.pyplot as plt

plt.bar([0, 1, 2], [3, 5, 7])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(0)
heights = rng.uniform(0, 100, 10)

plt.bar(np.arange(len(heights)), heights)
plt.show()
```
